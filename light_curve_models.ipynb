```python
%matplotlib inline
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import lightkurve as lk

import warnings 
warnings.filterwarnings('ignore')
```

```python
exoplanet_pred = pd.read_csv('exoplanet_predictions.csv').drop('Unnamed: 0',axis=1)
```

```python
exoplanet_pred.head()
```

PREPROCESSING DATA

```python
# We're more interested in the CONFIRMED predictions. Sort the dataframe of CONFIRMED prediction
# values by disposition scores, descending in value.

pred = exoplanet_pred['Predictions'].values

cand_pred_idx = np.where(pred == 'Pred CONFIRMED')

cand_data = exoplanet_pred.iloc[cand_pred_idx]
cand_data = cand_data.sort_values(by=['Disposition Score'],ascending=False)
cand_kepID = cand_data['KepID'].to_list()
```

PLOTTING CANDIDATE LIGHT CURVES

```python
# As an example, we'll analyze a known exoplanet first.

def light_curve(dso, start, end):
    id = str(dso)
    comb_curve_data = lk.search_lightcurve(id, author="Kepler", cadence="long")
    target_curve_data = comb_curve_data[start-1:end]
    tcd_stitched = target_curve_data.download_all().stitch()
    tcd_stitched.plot(column='pdcsap_flux')
```

```python
light_curve('Kepler-8',2,2)
```

```python
# For 50 of the first stars with data, we'll plot the light curves, and try to notice any tell-tale signs of transit.

for dso in range(50):
    light_curve(cand_kepID[dso],1,1)
```

INVESTIGATING KIC 5972334

```python
# From our plots, we see evidence of Kepler-487 b (a confirmed exoplanet), but not of Kepler-487c (a candidate exoplanet).

light_curve('5972334',1,5)
```

INVESTIGATING KIC 10656508

```python
# From our plots, we see evidence of clear and noticeable dips at BJKD 136, 508, 632, 756, 880, 1005, 
# 1128, 1252, 1376, 1500. I cannot personally determine if these transits are due to exoplanets, but 
# it seems like there is some good evidence for that argument.

light_curve('10656508',1,25)
```
